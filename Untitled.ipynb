```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [2,3,4,5,6]
y1 = [6.55, 4.43, 3.19, 3.19, 3.19]
y2 = [9.49, 6.44, 5.04, 5.04, 5.04]
```

```python
plt.plot(x, y1, "o-", label = "slider")
plt.plot(x, y2, "o-", label = "valve")

plt.xticks(np.arange(2, 7, 1))
plt.xlabel("The number of sources in a mixture", fontsize=12, labelpad=10)
plt.ylabel("SDR", fontsize=12, labelpad=10)
```
